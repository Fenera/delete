```python
#PACMAN CLASS

import pygame
import time

clock = pygame.time.Clock()

class Pacman:

    def __init__(self, screen, position, facing, dotcount):                       
                                        
        self.facing = facing        # the direction pacman is facing
        self.dotcount = dotcount    # number of dots remaining to complete the level
        self.screen = screen        #graphical display that will have everything on it
        self.og_position = list(position) #change position from (x,y) -> immutable TO [x,y] -> mutable
        self.position = self.og_position.copy()
        self.speed = 1.5 # speed of pacman
        self.og_speed = self.speed
        self.boost_time_end = 0
        self.is_boost = False

        # [0 = teleport left, 1 = teleport right]
        self.teleport_list = [False, False]

        #self.image is used to represent Pacman, which is a surface object
        self.image = pygame.Surface((GRID_SIZE, GRID_SIZE))
        self.image.fill(PACMAN_YELLOW)  #set color of pacman

        #self.rect is used to represent the position and dimensions of Pacman in terms of coordinates
        self.rect = self.image.get_rect(center = (self.position[0] - PTS, self.position[1] - PTS))

        self.width_window, self.height_window = self.screen.get_size() #get the dimensions of screen

        # param: pygame.Rect(top-left, width, height)
        # rects used for collision to determine if the user can teleport or not
        self.left_teleport = pygame.Rect((-4, GRID_SIZE * 14.5), (10, GRID_SIZE*2))
        self.right_teleport = pygame.Rect((GRID_SIZE * 29.8, GRID_SIZE * 14.5), (10, GRID_SIZE * 2))

        # rects used to prevent pacman from getting in an infinite loop of teleporting between the 2 sides
        self.return_left_teleport = pygame.Rect((GRID_SIZE + PTS, GRID_SIZE*14.5), (10, GRID_SIZE * 2))
        self.return_right_teleport = pygame.Rect((GRID_SIZE * 28, GRID_SIZE * 14.5), (10, GRID_SIZE * 2))


    def draw(self):
        #draws the image onto the window 
        self.screen.blit(self.image, (self.position[0], self.position[1]))

    # for debugging purpose
    def show_position(self):
        return self.og_position.copy()
    
    def reset_position(self):
        self.position = self.og_position.copy()
    
    def boost(self, duration = 6):
        self.is_boost = True
        self.speed = self.og_speed * 2
        self.boost_time_end = time.time() + duration

    def update_speed(self):
        if self.is_boost and time.time() > self.boost_time_end:
            self.is_boost = False
            self.speed = self.og_speed



    #facing: 0 = Left, 1 = Up, 2 = Right, 3 = Down
    def move(self):
        x_move = 0  # horizontal movement
        y_move = 0  # vertical movement
        keys = pygame.key.get_pressed()

        # Determine the direction Pacman should move
        if (keys[pygame.K_LEFT] or keys[pygame.K_a]):
            x_move -= self.speed
            self.facing = 0
        elif (keys[pygame.K_UP] or keys[pygame.K_w]):
            y_move -= self.speed
            self.facing = 1
        elif (keys[pygame.K_RIGHT] or keys[pygame.K_d]):
            x_move += self.speed
            self.facing = 2
        elif (keys[pygame.K_DOWN] or keys[pygame.K_s]):
            y_move += self.speed
            self.facing = 3


        #is_on_path parameters: is_on_path(pos, p1, p2, tolerance=5)

        # Only proceed with position updates if movement occurs
        if x_move != 0 or y_move != 0:
            # Check if the new position is on a valid path
            for path in path_list:
                for i in range(len(path) - 1):
                    if is_on_path((self.position[0] + x_move, self.position[1] + y_move), path[i], path[i + 1], tolerance=5):
                        self.position[0] += x_move
                        self.position[1] += y_move

                        # Update Pacman's rectangle to reflect the new position
                        self.rect.center = (self.position[0] + GRID_SIZE // 2, self.position[1] + GRID_SIZE // 2)
                        return  # Exit after moving along the path


    def teleport(self):
        # UNCOMMENT TO VISUALIZE TELEPORT RECTS

        #pygame.draw.rect(self.screen, red, self.left_teleport)
        #pygame.draw.rect(self.screen, red, self.right_teleport)
        #pygame.draw.rect(self.screen, green, self.return_left_teleport)
        #pygame.draw.rect(self.screen, green, self.return_right_teleport)

        # teleport_list[0 = LEFT, 1 = RIGHT]
        # check for collisions with the return rects(pacman is about to teleport but has not yet)
        if self.rect.colliderect(self.return_left_teleport):
            self.teleport_list[0] = True
            self.teleport_list[1] = False
        
        elif self.rect.colliderect(self.return_right_teleport):
            self.teleport_list[0] = False
            self.teleport_list[1] = True

        # check for collisions between the teleport rects and pacman rect
        # use teleport_list to prevent glitchy teleportation(0 and 1 are opposite booleans-- like 2 on/off light switches)
        # if left teleport is used, right teleport is off until the user moves forward a little(vice versa)
        if self.rect.colliderect(self.left_teleport) and self.teleport_list[0] == True and self.teleport_list[1] == False:
            self.position[0] = (GRID_SIZE * 29)
            self.position[1] = GRID_SIZE * 15
        elif self.rect.colliderect(self.right_teleport) and self.teleport_list[0] == False and self.teleport_list[1] == True:
            self.position[0] = 0
            self.position[1] = GRID_SIZE * 15
        


    def ghost_collision(self, rect):
        '''Checks if pacman collides with the ghost's collision rectangle, and if it does, it returns True. Otherwise, it returns False'''
        #note: rect is the ghost rect not pacman rect

        if self.rect.colliderect(rect):
            return True
        
        return False


    # pacman death and revive code is in ghost class(because any ghost can kill him so it's better to handle death/revive it in the ghost class)
```

```python
#GHOST CLASS

import random
import time

class Ghost:

    def __init__(self, screen, position, target, scattertarget, color, score):
            
        self.screen = screen    # graphical window
        self.original_position = list(position)
        self.position = self.original_position.copy() # store a copy of the og position
        self.facing = 0 # direction ghost is facing initially
        self.score = score # score object used to update the score
        self.target = (target[0]*GRID_SIZE, target[1]*GRID_SIZE) # ghost's target position(tuple)
        self.scattertarget = (scattertarget[0]*GRID_SIZE, scattertarget[1]*GRID_SIZE) # The ghost's specific "corner" in Scatter
        self.original_color = color # original color of the ghost
        self.speed = 1
        self.original_speed = 1
        self.image = pygame.Surface((GRID_SIZE, GRID_SIZE))
        self.image.fill(color)
        self.rect = self.image.get_rect(topleft = (self.position[0], self.position[1]))
        self.original_target = target # original target of the ghost

        # get width and height of screen(used in move())
        self.width_screen, self.height_screen = screen.get_size()

        self.scared_until = 0 # used to store a time representing how long the ghost is in a scared state
        self.scared_state = False # used to store boolean when the ghost is/not in a scared state(blue)

    def draw(self):
        self.screen.blit(self.image, (self.position[0], self.position[1]))

    def set_color(self, color, duration=0):
        '''changes the color of the ghosts'''
        self.color = color
        self.image.fill(pygame.Color(color))
        if duration > 0:
            self.scared_until = time.time() + duration
    
    def generate_avoid_target(self):
        '''create a random target for the ghost to go to when it is in a scared state'''
        xpos, ypos = random.randint(0, ncol - 1), random.randint(0, nrow - 1)
        return (xpos * GRID_SIZE, ypos * GRID_SIZE)


    def set_scared_state(self, scared_duration):
        '''sets the scared state for the ghost when it pacman consumes a large pellet'''
        self.scared_state = True # ghost is in a scared state
        self.scared_time_end = time.time() + scared_duration
    
    def update(self):
        '''reverts the ghosts color and target once the scared state has ended'''
        # if scared time is over, revert target and color
        if self.scared_state and time.time() > self.scared_until: # the current time is greater than the time the ghost should be scared(time expired)
            self.image.fill(self.original_color)
            self.target = self.original_target
            self.scared_state = False
    
    def eaten(self, pacman):
        # checks if the ghost has been eaten by pacman while in scared state
        if self.scared_state:
            if self.rect.colliderect(pacman.rect):
                self.score.update(200)
                self.position = (GRID_SIZE * 15, GRID_SIZE * 15)

    def eat_pacman(self, pacman):
        '''handles the instances where pacman is eaten by ghost'''
        if not self.scared_state: # if ghost is not blue(scared state)
            if self.score.get_score() > 0: # at least 1 life left
                if self.rect.colliderect(pacman.rect):
                    self.score.decrease_lives() # -1 life
                    pygame.time.wait(3500) # wait 3.5 seconds before resetting position
                    self.position = self.original_position
                    pacman.reset_position()
            else:
                # pacman has 0 lives so program terminates
                #pygame.quit()
                pass



    def face_target(self):
        '''Calculates how the ghost should face to reach their target
        on an intersection of grid lines.'''
        if self.position[0]%GRID_SIZE == 0 and (self.position[1]+TOP_MARGIN)%GRID_SIZE == 0:
            self.facing = self.turncalc()

    def set_target(self, x, y):
        """Sets the target position for the ghost."""
        if self.scared_state: # check if the ghost is in a scared state
            self.target = self.generate_avoid_target() # set target to random target 
        else:
            self.target = (x, y)    
    
    def turncalc(self):
        '''Finds the best direction to move toward the target while avoiding walls.'''
        best_distance = float('inf')
        best_direction = None

        for direction in range(4):  # Check all directions
            if direction != (self.facing + 2) % 4:  # Avoid 180° turns
                test_distance = self.distancecalc(direction)
                if test_distance < best_distance:
                    best_distance = test_distance
                    best_direction = direction

        return best_direction if best_direction is not None else self.facing

    def distancecalc(self, direction):
        '''Checks distance from target after moving one tile in the specified direction.
        Ensures the move is valid.'''
        testx, testy = self.position[0], self.position[1]

        if direction == 0:  # Left
            testx -= GRID_SIZE
        elif direction == 1:  # Up
            testy -= GRID_SIZE
        elif direction == 2:  # Right
            testx += GRID_SIZE
        elif direction == 3:  # Down
            testy += GRID_SIZE

        # Check if the new position is valid
        if not self.is_valid_position((testx, testy)):
            return float('inf')  # Invalid moves are heavily penalized

        # Calculate Euclidean distance to target
        return ((testx - self.target[0])**2 + (testy - self.target[1])**2)**0.5
    
    def is_valid_position(self, position):
        '''Checks if the given position is valid (not a wall).'''
        x, y = position
        if x < 0 or y < 0 or x >= self.width_screen or y >= self.height_screen:
            return False
        return any(
            is_on_path(position, path[i], path[i + 1])
            for path in ghost_path_list
            for i in range(len(path) - 1)
        )

    def is_at_intersection(self):
        '''This code checks if the ghost is at an intersection which is a point directly between 2 cells(at the grid line)
            If the ghost is at an intersection, the function returns True, else False'''
        return self.position[0] % GRID_SIZE == 0 and (self.position[1]) % GRID_SIZE == 0
    
    def avoid_stuck_teleport(self, teleport_left_rect, teleport_right_rect):
        '''
        The ghosts sometimes get stuck by the teleporters so this function changes the direction so they can avoid doing that
        '''

        # directions for self.facing: 0 = left, 1 = up, 2 = right, 3 = down

        # check if the teleporter is the left
        if self.rect.colliderect(teleport_left_rect):
            #change the direction the ghost should face to right
            self.facing = 2
        
        # check if the teleporter is the left
        if self.rect.colliderect(teleport_right_rect):
            #change the ghost's direction to face the opposite way(left)
            self.facing = 0
    
    def change_direction(self):
        # Try new directions in random order until a valid one is found
        directions = list(range(4))
        random.shuffle(directions)  # Randomize order of attempts

        for direction in directions:
            if direction != (self.facing + 2) % 4:  # Avoid 180° turns
                self.facing = direction
                dx, dy = 0, 0
                if direction == 0:  # Left
                    dx -= self.speed
                elif direction == 1:  # Up
                    dy -= self.speed
                elif direction == 2:  # Right
                    dx += self.speed
                elif direction == 3:  # Down
                    dy += self.speed

                new_position = (self.position[0] + dx, self.position[1] + dy)

                # Check if the new position is valid
                if any(
                    is_on_path(new_position, path[i], path[i + 1])
                    for path in ghost_path_list
                    for i in range(len(path) - 1)
                ):
                    return  # Valid direction found
                 
    def move(self):
        dx, dy = 0, 0

        # Recalculate direction at intersections(no walls around and can decide by itself on best direction)
        if self.is_at_intersection():
            self.facing = self.turncalc()

        # Update movement based on facing direction
        if self.facing == 0 and self.position[0] > 0:  # Left
            dx -= self.speed
        elif self.facing == 1 and self.position[1] > 0:  # Up
            dy -= self.speed
        elif self.facing == 2 and self.position[0] < self.width_screen - GRID_SIZE:  # Right
            dx += self.speed
        elif self.facing == 3 and self.position[1] < self.height_screen - GRID_SIZE:  # Down
            dy += self.speed

        new_position = (self.position[0] + dx, self.position[1] + dy)

        # Validate the new position
        if self.is_valid_position(new_position):
            self.position = list(new_position)
            self.rect.topleft = self.position
        else:
            self.change_direction()


        
    def blinkytarget(self, pacx, pacy):
        '''Blinky directly targets Pacman's position.'''
        self.set_target(pacx, pacy)

    def pinkytarget(self,pacx,pacy,pacfacing):
        '''Pinky tries to get in front of pacman.'''
        out_x = (pacx//GRID_SIZE+1)
        out_y = (pacy//GRID_SIZE+1)
        if pacfacing == 0:
            out_x -= 4
        elif pacfacing == 1:
            out_y -= 4
        elif pacfacing == 2:
            out_x += 4
        elif pacfacing == 3:
            out_y += 4
        out_x *= GRID_SIZE
        out_y *= GRID_SIZE
        self.set_target(out_x,out_y)

    def inkytarget(self,partnerx,partnery,pacx,pacy,pacfacing):
        '''Inky attempts to pincer with Blinky.'''
        partnerx = (partnerx//GRID_SIZE+1)
        partnery = (partnery//GRID_SIZE+1)
        pacx = (pacx//GRID_SIZE+1)
        pacy = (pacy//GRID_SIZE+1)
        if pacfacing == 0:
            pacx -= 2
        elif pacfacing == 1:
            pacy -= 2
        elif pacfacing == 2:
            pacx += 2
        elif pacfacing == 3:
            pacy += 2
        deltax = pacx - partnerx
        deltay = pacy - partnery
        out_x = (partnerx + deltax*2)*GRID_SIZE
        out_y = (partnery + deltay*2)*GRID_SIZE
        self.set_target(out_x,out_y)

    def clydetarget(self,pacx,pacy):
        '''Clyde acts like Blinky until pacman is too close.
        In that case, he targets his scatter corner.'''
        out_x = (pacx//GRID_SIZE+1)*GRID_SIZE
        out_y = (pacy//GRID_SIZE+1)*GRID_SIZE

        if self.scared_state:
            self.target = self.generate_avoid_target() # set target to random target 
        else:
            self.target = ((out_x,out_y))

        if self.distancecalc(-1) <= GRID_SIZE*8:
            
            if self.scared_state:
                self.target = self.generate_avoid_target() # set target to random target 
            else:
                self.target = self.scattertarget

    def scatter(self):
        self.facing = (self.facing+2)%4
        self.set_target(self.scattertarget[0],self.scattertarget[1])
```

```python
#PELLET CLASS
import pygame
import time

class Pellet:

    def __init__(self, screen, posx, posy, radius, score):

        #set attributes of Pellet

        self.screen = screen #graphical window
        self.posx = posx     #x position of pellet
        self.posy = posy     #y position of pellet
        self.radius = radius #radius of pellet
        self.score = score


        self.color = white
        self.is_visible = True

    def draw(self):
        '''
        draws the pellet on the window if it should be currently visible(not eaten by Pacman)
        '''

        if self.is_visible:
            
            #calculate the width and height of the pellet rect
            self.rect_width, self.rect_height = self.radius * 2, self.radius * 2

            #create a pellet rect to check for collisions with other rects(such as pacman)
            self.pellet_rect = pygame.Rect(self.rect_width - self.radius, self.rect_height - self.radius, self.rect_width, self.rect_height)

            transparency_level = 0 #0 = 100% transparent, 255 = 0% transparent/fully solid

            #create a transparent surface(using the alpha channel)
            transparent_surface = pygame.Surface((self.rect_width, self.rect_height), pygame.SRCALPHA)
            transparent_surface.set_alpha(transparency_level)#set the alpha channel to the transparency level

            #draw the pellet_rect on the transparent surface(which makes the pellet_rect also transparent)
            pygame.draw.rect(transparent_surface, white, self.pellet_rect)

            #draw the pellet
            pygame.draw.circle(self.screen, white, (self.posx, self.posy), self.radius)
    
    def collide_pacman(self, pacman, ghosts):
        '''
        checks if pacman has eaten the pellet(collided with it)
            
        if it has been eaten, it takes the pellet off the visible part of the window(shown by a negative coordinate)
        '''
    
        # Update the pellet_rect to use current position (self.posx, self.posy) and radius
        self.pellet_rect = pygame.Rect(self.posx - self.radius, self.posy - self.radius, self.radius * 2, self.radius * 2)
        
        #check if pacman's rect has collided with the pellet's rect
        if self.pellet_rect.colliderect(pacman.rect):
            
            #check if pacman collided with a large pellet or a small pellet

            if self.radius == 5: #small pellet
                self.score.update(10) # increase score by 10

                
                
            elif self.radius == 10: #large pellet
                self.score.update(50) # increase score by 50
                pacman.boost() # increase speed of player if they eat energizer pellet
                #turn all ghosts blue
                for ghost in ghosts:
                    ghost.set_color(SCARED_GHOST, duration = 6)
                    ghost.set_scared_state(6)
                    

            #change the x and y position of the pellet
            self.posx = -10
            self.posy = -10

            #draw the pellet off-screen so it gives the illusion it is no longer visible
            pygame.draw.circle(self.screen, white, (self.posx, self.posy), self.radius)

            self.is_visible = False #set to False because this pellet should no longer be drawn
```

```python
class Score:

    def __init__(self):
        self.total_score = 0
        self.lives = 3

    def update(self, points_earned):
        '''updates the user's score'''
        self.total_score += points_earned
    
    def get_score(self):
        '''returns the current total score'''
        return self.total_score

    def decrease_lives(self):
        self.lives -= 1

    def get_lives(self):
        '''returns the number of lives the pacman has'''
        return self.lives

    def reset(self):
        '''use if we want 'restart game' functionality'''
        self.total_score = 0
```
