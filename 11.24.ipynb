```python
import heapq

path1 = [(GRID_SIZE * 3, GRID_SIZE * 3), (GRID_SIZE * 14, GRID_SIZE * 3), (GRID_SIZE * 14, GRID_SIZE * 7),(GRID_SIZE * 3, GRID_SIZE * 7),(GRID_SIZE * 3, GRID_SIZE * 3)] #1-21-22-25-1        
path2 = [(GRID_SIZE * 14, GRID_SIZE * 3), (GRID_SIZE * 28, GRID_SIZE * 3), (GRID_SIZE * 28, GRID_SIZE * 7),(GRID_SIZE * 14, GRID_SIZE * 7),(GRID_SIZE * 14, GRID_SIZE * 3)]#21-2-29-22-21
path3 = [(GRID_SIZE * 14, GRID_SIZE * 3), (GRID_SIZE * 14, GRID_SIZE * 7),(GRID_SIZE * 17, GRID_SIZE * 7), (GRID_SIZE * 17, GRID_SIZE * 3), (GRID_SIZE * 14, GRID_SIZE * 3)]#21-22-23-24-21
path4 = [(GRID_SIZE * 3, GRID_SIZE * 7), (GRID_SIZE * 8, GRID_SIZE * 7), (GRID_SIZE * 8, GRID_SIZE * 10), (GRID_SIZE * 3, GRID_SIZE * 10),(GRID_SIZE * 3, GRID_SIZE * 7)]#25-28-27-27-25
path5 = [(GRID_SIZE * 3, GRID_SIZE * 10), (GRID_SIZE * 8, GRID_SIZE * 10), (GRID_SIZE * 8, GRID_SIZE * 13), (GRID_SIZE * 3, GRID_SIZE * 13),(GRID_SIZE * 3, GRID_SIZE * 10)]#26-27-19-20-26            
path6 = [(GRID_SIZE * 28, GRID_SIZE * 7), (GRID_SIZE * 28, GRID_SIZE * 10), (GRID_SIZE * 23, GRID_SIZE * 10), (GRID_SIZE * 23, GRID_SIZE * 7), (GRID_SIZE * 28, GRID_SIZE * 7)]#29-30-31-32-29             
path7 = [(GRID_SIZE * 23, GRID_SIZE * 10), (GRID_SIZE * 28, GRID_SIZE * 10), (GRID_SIZE * 28, GRID_SIZE * 13), (GRID_SIZE * 23, GRID_SIZE * 13),(GRID_SIZE * 23, GRID_SIZE * 10)]#31-30-3-4-31             
path8 = [(GRID_SIZE * 8, GRID_SIZE * 19), (GRID_SIZE * 3, GRID_SIZE * 19), (GRID_SIZE * 3, GRID_SIZE * 22), (GRID_SIZE * 8, GRID_SIZE * 22),(GRID_SIZE * 8, GRID_SIZE * 19)]#18-17-50-43-18
path9 = [(GRID_SIZE * 3, GRID_SIZE * 22), (GRID_SIZE * 3, GRID_SIZE * 25), (GRID_SIZE * 5, GRID_SIZE * 25), (GRID_SIZE * 5, GRID_SIZE * 28), (GRID_SIZE * 3, GRID_SIZE * 28), (GRID_SIZE * 3, GRID_SIZE * 31), (GRID_SIZE * 14,  GRID_SIZE * 31), (GRID_SIZE * 14, GRID_SIZE * 28), (GRID_SIZE * 11, GRID_SIZE * 28), (GRID_SIZE * 11, GRID_SIZE * 25), (GRID_SIZE * 8, GRID_SIZE * 25), (GRID_SIZE * 8, GRID_SIZE * 28), (GRID_SIZE * 5, GRID_SIZE * 28), (GRID_SIZE * 5, GRID_SIZE * 25), (GRID_SIZE * 3, GRID_SIZE * 25), (GRID_SIZE * 3, GRID_SIZE * 22)]#50-16-15-14-13-12-49-48-47-46-44-45-14-15-16-50
path10 = [(GRID_SIZE * 8, GRID_SIZE * 22), (GRID_SIZE * 14, GRID_SIZE * 22), (GRID_SIZE * 14, GRID_SIZE * 25), (GRID_SIZE * 8, GRID_SIZE * 25), (GRID_SIZE * 8, GRID_SIZE * 22)] #43-50-17-18-43
path11 = [(GRID_SIZE * 8, GRID_SIZE * 19), (GRID_SIZE * 8, GRID_SIZE * 13)]
path12 = [(GRID_SIZE * 17, GRID_SIZE * 31), (GRID_SIZE * 28, GRID_SIZE * 31), ( GRID_SIZE * 28, GRID_SIZE * 31), (GRID_SIZE * 28, GRID_SIZE * 28), (GRID_SIZE * 26, GRID_SIZE * 28), (GRID_SIZE * 23, GRID_SIZE * 28), (GRID_SIZE * 23, GRID_SIZE * 25), (GRID_SIZE * 20, GRID_SIZE * 25), (GRID_SIZE * 20, GRID_SIZE * 28), (GRID_SIZE * 17, GRID_SIZE * 28)] #33-11-10-9-58-37-36
path13 = [(GRID_SIZE * 14, GRID_SIZE * 25), (GRID_SIZE * 20, GRID_SIZE * 25)]
path14 = [(GRID_SIZE * 17, GRID_SIZE * 25), (GRID_SIZE * 17, GRID_SIZE * 22), (GRID_SIZE * 23, GRID_SIZE * 22), (GRID_SIZE * 23, GRID_SIZE * 25), (GRID_SIZE * 17, GRID_SIZE * 25)]#40-39-38-37-40
path15 = [(GRID_SIZE * 23, GRID_SIZE * 22), (GRID_SIZE * 23, GRID_SIZE * 19), (GRID_SIZE * 28, GRID_SIZE * 19), (GRID_SIZE * 28, GRID_SIZE * 22), (GRID_SIZE * 23, GRID_SIZE * 22)]#38-5-6-51-38
path16 = [(GRID_SIZE * 28, GRID_SIZE * 22), (GRID_SIZE * 28, GRID_SIZE * 25), (GRID_SIZE * 26, GRID_SIZE * 25), (GRID_SIZE * 26, GRID_SIZE * 28),(GRID_SIZE * 23, GRID_SIZE * 28),(GRID_SIZE * 23, GRID_SIZE * 22),(GRID_SIZE * 28, GRID_SIZE * 22)] #51-7-8-9-58-38-51
path17 = [(GRID_SIZE * 23, GRID_SIZE * 13), (GRID_SIZE * 23, GRID_SIZE * 19),(GRID_SIZE * 23, GRID_SIZE * 13)]          #a                              #b                              #c                              #d                                  #e                                      #f                                     #g                          #h
path_mid1 = [(GRID_SIZE * 11, GRID_SIZE * 7), (GRID_SIZE * 11, GRID_SIZE *  10), (GRID_SIZE * 14, GRID_SIZE * 10), (GRID_SIZE * 14, GRID_SIZE * 13), (GRID_SIZE * 11, GRID_SIZE * 13), (GRID_SIZE * 11, GRID_SIZE * 16), (GRID_SIZE * 11, GRID_SIZE * 19), (GRID_SIZE * 20, GRID_SIZE * 19), (GRID_SIZE * 20, GRID_SIZE * 16), (GRID_SIZE * 20, GRID_SIZE * 13), (GRID_SIZE * 17, GRID_SIZE * 13), (GRID_SIZE * 17, GRID_SIZE * 10), (GRID_SIZE * 20, GRID_SIZE * 10), (GRID_SIZE * 20, GRID_SIZE * 7), (GRID_SIZE * 11, GRID_SIZE * 7)]#52-53-54-a-b-c-d-e-f-g-h-57-56-55-52
left_teleport = [(GRID_SIZE * 1, GRID_SIZE * 16), (GRID_SIZE * 8, GRID_SIZE * 16), (GRID_SIZE * 11, GRID_SIZE * 16)]
right_teleport = [(GRID_SIZE * 30, GRID_SIZE * 16), (GRID_SIZE * 23, GRID_SIZE * 16), (GRID_SIZE * 20, GRID_SIZE * 16)]
a_h = [(GRID_SIZE * 14, GRID_SIZE * 13), (GRID_SIZE * 17, GRID_SIZE * 13)]
d_below = [(GRID_SIZE * 11, GRID_SIZE * 13), (GRID_SIZE * 11, GRID_SIZE * 22)]
e_below = [(GRID_SIZE * 20, GRID_SIZE * 19), (GRID_SIZE * 20, GRID_SIZE * 22)]
f9_t3_con = [(GRID_SIZE * 14,  GRID_SIZE * 31), (GRID_SIZE * 17, GRID_SIZE * 31)]
path18 = [(GRID_SIZE * 17, GRID_SIZE * 31), (GRID_SIZE * 17, GRID_SIZE * 28)]
#33-34
path_list = [path1, path2, path3, path4, path5, path6, path7, path8, path9, path10, path11, path12, path13, path14, path15, path16, path17,path18, path_mid1, left_teleport, right_teleport, a_h, e_below, d_below, f9_t3_con]

#off_set path_list by (x-PTS, y-PTS) to center it in the cell
for index, path in enumerate(path_list): #enumerate allows us to get both the index and the path in one iteration

    path_list[index] = [(x - GRID_SIZE, y - GRID_SIZE) for x,y in path]

def draw_path(screen):
    for path in path_list:
        # Draw each point (circle) in the offset_rect
        for point in path:
            pygame.draw.circle(screen, green, point, 12)  # Draw circles at each point
            
        # Draw the lines connecting the points in offset_rect
        pygame.draw.lines(screen, green, True, path, 2)  # True to close the shape


def is_on_path(pos, p1, p2, tolerance=PTS):
    #checks to see if sprite is close enough to the path to be considered on it
    #p1 is starting point (x,y), p2 is ending point (x,y)

    if p1[0] == p2[0]:#if x's are the same, line is vertical
        return min(p1[1], p2[1]) - tolerance <= pos[1] <= max(p1[1], p2[1]) + tolerance and abs(pos[0] - p1[0]) <= tolerance #checks if the character is within the vertical tolerance area
    elif p1[1] == p2[1]:#if y's are the same, line is horizontal
        return min(p1[0], p2[0]) - tolerance <= pos[0] <= max(p1[0], p2[0]) + tolerance and abs(pos[1] - p1[1]) <= tolerance#checks if the sprite is within the horizontal tolerance area
    
    return False 


#guide on the A* algorithm: 
# https://www.youtube.com/watch?v=i0x5fj4PqP4
# https://www.geeksforgeeks.org/a-search-algorithm/

def get_traversible_points(pathList):
    '''
    converts coordinates from: (GRID_SIZE * 17, GRID_SIZE * 31) => (17, 31)

    return all traversable points as a set(unique elements)
    '''

    traversable_points = set()

    for path in pathList: 
        traversable_points.add(tuple((x // GRID_SIZE, y // GRID_SIZE) for x, y in path))
    return traversable_points

def calc_manhattan_distance(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])  


def A_StarAlgorithm(map_list, start, goal):
    '''
    function: find the best way to reach the target(end) cell from the start cell
    map_list: list of points defining places where sprites can move
    start: start position
    goal: goal/target position
                    
               For square and rectangular cells, neighboring cells are defined as the following: 

                              (0, -1)
                                 ↑
                   (-1, 0) ← current(x,y) → (1, 0)
                                ↓
                               (0, 1)
                
    '''
    map_list = list(map_list)
    print(map_list)
    rows, cols = len(map_list), len(map_list[0])

    open_set = [] # priority queue
    heapq.heappush(open_set, (0, start)) # Add the start node with initial priority 
    came_from = {} # dictionary to track the path
    g_score = {start: 0} # store g-cost from start to each node
    f_score = {start: calc_manhattan_distance(start, goal)} # store total estimated cost for each node 

    # loop until no nodes left to explore
    while open_set:

        _, current = heapq.heappop(open_set) # get the node with the lowest f-cost

        if current == goal: # If the goal is reached, reconstruct the path 

            path = []

            while current in came_from: # trace back from goal to start
                path.append(current)
                current = came_from[current]

            path.reverse() # reverse the path to start from the beginning

            return path
        
        x, y = current

        for dx, dy  in [(-1, 0), (1, 0), (0, -1), (0, 1)]: # check all adjacent neighbors
            neighbor = (x + dx, y + dy) # calculate neighbors x and y position
        
            # check if neighbor is valid and within grid bounds
            if neighbor in map_list:
                if map_list[neighbor[0][neighbor[1]]] is None: # we don't have none so avoid this
                    continue

                # calculate neighbor g-cost/g-cost
                neighbor_g = g_score[current] + 1

                # update path if this path to neighbor is better
                if neighbor not in g_score or neighbor_g < g_score[neighbor]:
                    came_from[neighbor] = current 
                    g_score[neighbor] = neighbor_g # update g-score/g-cost for neighbor
                    f_score[neighbor] = neighbor_g + calc_manhattan_distance(neighbor, goal) # update f-score/f-cost

                    # add neighbor to open set if it is not already there
                    if neighbor not in map_list:
                        heapq.heappush(open_set, (f_score[neighbor], neighbor))

    return [] # return empty list if no path found


# test with example map

test_map = get_traversible_points(path_list)
print(test_map)

starting_point = (7, 27)
target_point = (25, 24)

A_StarAlgorithm(test_map, starting_point, target_point)














```

```python
#GHOST CLASS

import random

class Ghost:

    def __init__(self, screen, position, facing, target, scattertarget, color):
            
        self.screen = screen    #graphical window
        self.position = list(position)
        self.facing = 0 #direction ghost is facing
        self.target = (target[0]*GRID_SIZE, target[1]*GRID_SIZE)#ghost's target position(tuple)
        self.scattertarget = (scattertarget[0]*GRID_SIZE, scattertarget[1]*GRID_SIZE) # The ghost's specific "corner" in Scatter

        self.speed = 1
        self.image = pygame.Surface((GRID_SIZE, GRID_SIZE))
        self.image.fill(color)
        self.rect = self.image.get_rect(topleft = (self.position[0], self.position[1]))

        # [left, top, right, bottom]
        self.direction_facing = [0, 1, 2, 3]

        #get width and height of screen(used in move())
        self.width_screen, self.height_screen = screen.get_size()


    def draw(self):
        self.screen.blit(self.image, (self.position[0], self.position[1]))
    

    def face_target(self):
        '''Calculates how the ghost should face to reach their target
        on an intersection of grid lines.'''
        if self.position[0]%GRID_SIZE == 0 and (self.position[1]+TOP_MARGIN)%GRID_SIZE == 0:
            self.facing = self.turncalc()
    
    def turncalc(self):
        '''Find the best direction to turn based on the ghost's target.
        Minimizes geometric distance from target; no other criteria.
        Cannot turn 180 degrees.'''
        # print("calling turncalc")
        best = 1000
        best_direction = None
        if best_direction is None:
            best_direction = 0
        for direction in range(0,4):
            if direction != (self.facing+2)%4:
                # print(f"checking direction {direction}")
                testdist = self.distancecalc(direction)
                if testdist < best:
                    best = testdist
                    best_direction = direction
                    # print(f"new best is {best_direction}")
        return best_direction
    
    def turncalc_random(self):
        '''Chooses a valid turn direction at random.'''
        while True:
            direction = random.randrange(0,4)
            if direction != (self.facing+2)%4:
                return direction


    def distancecalc(self,direction):
        '''Checks distance from target after moving one tile in the specified direction.
        If a value other than 0-3 is provided, uses current location.'''
        # print("Calling distancecalc")
        testx = self.position[0]
        testy = self.position[1]
        if direction == 0:
            testx -= GRID_SIZE
        elif direction == 1:
            testy -= GRID_SIZE
        elif direction == 2:
            testx += GRID_SIZE
        elif direction == 3:
            testy += GRID_SIZE
        distance = (abs(testx-self.target[0])**2+abs(testy-self.target[1])**2)**(1/2)
        # print(distance)
        return distance
    
    def set_target(self,posx,posy):
        self.target = (posx,posx)

    def move(self):

        dx, dy = 0,0    # dx = horizontal movement, dy = vertical movement
        if self.facing == 0 and self.position[0] > 0:
            dx -= self.speed
        if self.facing == 1 and self.position[1] > 0:
            dy -= self.speed
        if self.facing == 2 and self.position[0] < self.width_screen - GRID_SIZE:
            dx += self.speed
        if self.facing == 3 and self.position[1] < self.height_screen - GRID_SIZE:
            dy += self.speed
        
        new_position = (self.position[0] + dx, self.position[1] + dy)


        # ensure the ghost stays on the path

        for path in path_list: 
            for i in range(len(path) - 1):

                if is_on_path(new_position, path[i], path[i + 1]):
                    self.position = list(new_position)
                    self.rect.topleft = self.position
                    return # exit after a valid path is found
                
        #if there is no valid path, snap ghost to nearest valid position on the current path
        self.snap_to_path()


    def snap_to_path(self):
        """
        Snap the ghost to the nearest path point.
        """
        closest_distance = float('inf')
        closest_point = None

        for path in path_list:
            for point in path:
                distance = ((self.position[0] - point[0])**2 + (self.position[1] - point[1])**2)**0.5
                if distance < closest_distance:
                    closest_distance = distance
                    closest_point = point

        if closest_point:
            self.position = list(closest_point)
            self.rect.topleft = self.position
            self.facing = random.choice(self.direction_facing)

    def blinkytarget(self,pacx,pacy):
        '''Blinky targets pacman directly.'''
        out_x = (pacx//GRID_SIZE+1)*GRID_SIZE
        out_y = (pacy//GRID_SIZE+1)*GRID_SIZE
        self.target = ((out_x,out_y))

    def pinkytarget(self,pacx,pacy,pacfacing):
        '''Pinky tries to get in front of pacman.'''
        out_x = (pacx//GRID_SIZE+1)
        out_y = (pacy//GRID_SIZE+1)
        if pacfacing == 0:
            out_x -= 4
        elif pacfacing == 1:
            out_y -= 4
        elif pacfacing == 2:
            out_x += 4
        elif pacfacing == 3:
            out_y += 4
        out_x *= GRID_SIZE
        out_y *= GRID_SIZE
        self.set_target(out_x,out_y)

    def inkytarget(self,partnerx,partnery,pacx,pacy,pacfacing):
        '''Inky attempts to pincer with Blinky.'''
        partnerx = (partnerx//GRID_SIZE+1)
        partnery = (partnery//GRID_SIZE+1)
        pacx = (pacx//GRID_SIZE+1)
        pacy = (pacy//GRID_SIZE+1)
        if pacfacing == 0:
            pacx -= 2
        elif pacfacing == 1:
            pacy -= 2
        elif pacfacing == 2:
            pacx += 2
        elif pacfacing == 3:
            pacy += 2
        deltax = pacx - partnerx
        deltay = pacy - partnery
        out_x = (partnerx + deltax*2)*GRID_SIZE
        out_y = (partnery + deltay*2)*GRID_SIZE
        self.set_target(out_x,out_y)

    def clydetarget(self,pacx,pacy):
        '''Clyde acts like Blinky until pacman is too close.
        In that case, he targets his scatter corner.'''
        out_x = (pacx//GRID_SIZE+1)*GRID_SIZE
        out_y = (pacy//GRID_SIZE+1)*GRID_SIZE
        self.target = ((out_x,out_y))
        if self.distancecalc(-1) <= GRID_SIZE*8:
            self.target = self.scattertarget

    def scatter(self):
        self.facing = (self.facing+2)%4
        self.set_target(self.scattertarget[0],self.scattertarget[1])
```

```python
#PELLET CLASS
import pygame

class Pellet:

    def __init__(self, screen, posx, posy, radius):

        #set attributes of Pellet

        self.screen = screen #graphical window
        self.posx = posx     #x position of pellet
        self.posy = posy     #y position of pellet
        self.radius = radius #radius of pellet


        self.color = white
        self.is_visible = True
        self.score = 0

    def draw(self):
        '''
        draws the pellet on the window if it should be currently visible(not eaten by Pacman)
        '''

        if self.is_visible:
            
            #calculate the width and height of the pellet rect
            self.rect_width, self.rect_height = self.radius * 2, self.radius * 2

            #create a pellet rect to check for collisions with other rects(such as pacman)
            self.pellet_rect = pygame.Rect(self.rect_width - self.radius, self.rect_height - self.radius, self.rect_width, self.rect_height)

            transparency_level = 0 #0 = 100% transparent, 255 = 0% transparent/fully solid

            #create a transparent surface(using the alpha channel)
            transparent_surface = pygame.Surface((self.rect_width, self.rect_height), pygame.SRCALPHA)
            transparent_surface.set_alpha(transparency_level)#set the alpha channel to the transparency level

            #draw the pellet_rect on the transparent surface(which makes the pellet_rect also transparent)
            pygame.draw.rect(transparent_surface, white, self.pellet_rect)

            #draw the pellet
            pygame.draw.circle(self.screen, white, (self.posx, self.posy), self.radius)
    
    def collide_pacman(self, pacman_rect):
        '''
        checks if pacman has eaten the pellet(collided with it)
        
        if it has been eaten, it takes the pellet off the visible part of the window(shown by a negative coordinate)
        '''
        
        # Update the pellet_rect to use current position (self.posx, self.posy) and radius
        self.pellet_rect = pygame.Rect(self.posx - self.radius, self.posy - self.radius, self.radius * 2, self.radius * 2)
        
        #check if pacman's rect has collided with the pellet's rect
        if self.pellet_rect.colliderect(pacman_rect):
            
            #check if pacman collided with a large pellet or a small pellet

            if self.radius == 5: #small pellet
                self.score += 10
            elif self.radius == 10: #large pellet
                self.score += 50

            #change the x and y position of the pellet
            self.posx = -10
            self.posy = -10

            #draw the pellet off-screen so it gives the illusion it is no longer visible
            pygame.draw.circle(self.screen, white, (self.posx, self.posy), self.radius)

            self.is_visible = False #set to False because this pellet should no longer be drawn
```
