```python
import pygame


clock = pygame.time.Clock()

class Pellet:
    '''
    Pellet class to make a single pellet object and draw it onto a window

    Parameters: 
    1. screen -> the display the pellets are drawn on
    2. posx -> the x-coordinate of the center of the pellet
    3. posy -> the y-coordinate of the center of the pellet
    '''
    
    def __init__(self, screen, posx, posy):
        self.screen = screen
        self.posx = posx
        self.posy = posy
        
        self.radius = 10
        self.color = white
        self.is_visible = True

    
    def draw(self):
        '''
        draws the pellet on the window if it should be visible(if it has not been eaten by pacman)
        '''
        if self.is_visible:
            #create a rectangle for collision purposes
            self.rect_width, self.rect_height = self.radius * 2, self.radius * 2
            self.pellet_rect = pygame.Rect(self.posx - self.radius, self.posy - self.radius, self.rect_width, self.rect_height)

            #draw a transparent surface
            transparency = 0 #100% transparent
            #creates a surface that can have transparency (alpha channel)
            transparent_surface = pygame.Surface((self.rect_width, self.rect_height), pygame.SRCALPHA)
            transparent_surface.set_alpha(transparency)

            #draw the collision rectangle onto the transparent surface(so it also becomes transparent)
            pygame.draw.rect(transparent_surface, white, self.pellet_rect)

            #draw the pellet
            pygame.draw.circle(self.screen, white, (self.posx, self.posy), self.radius)


    def collide_pacman(self, pacman_rect):
        '''
        checks if pacman has eaten the pellet(collided with it)
        
        if it has been eaten, it takes the pellet off the visible part of the window(explained by negative coordinate)
        '''
        #if pacman and the pellet collide
        if self.pellet_rect.colliderect(pacman_rect):
            
            #change the x and y value of the center of the pellet to negative numbers to move it off the screen
            self.posx = -10
            self.posy = -10

            #draw the pellet at the updated x and y position
            pygame.draw.circle(self.screen, white, (self.posx, self.posy), self.radius)
            
            #set the visible to False(should no longer be drawn)
            self.is_visible = False
        
        





class Ghost:
    '''
    creates a Ghost character that can be drawn onto a window
    For similarity to original game, it would probably be better to create a class for each of the 4 ghosts so their movements can be specified in their own classes
    for starters, I think it is fine to have a single class(we can complicate later)

    parameters:
    1.screen -> graphical window
    2. posx -> center of ghost(x)
    3. posy -> center of ghost(y)
    4. color -> color of ghost
    '''

    def __init__(self, screen, posx, posy, color, radius = 25, facing=0, image_path = ""):

        self.screen = screen    #window
        self.posx = posx        #x position of ghost object
        self.posy = posy        #y position of ghost object
        self.facing = facing    #direction the ghost is facing
        self.color = color      #color of ghost(blinky = red, pinky = pink, inky = cyan, clyde = orange)
        self.image_path = image_path  #image path of ghost(if we choose to import images from the web)
        self.radius = radius



    def draw(self):
        '''
        temporary ghost sprites(only used to check collisions)
            when we have actual sprites, the code can be modified to place them on the window
            '''
        #draw a rect around the circle
        #parameters: (left:x-coor. top left, top: y-coor. top left, width of rect, height of rect)
        self.rect_width = 2*self.radius
        self.rect_height = 2*self.radius
        self.ghost_rect = pygame.Rect(self.posx - self.radius, self.posy - self.radius, self.rect_width, self.rect_height)

        #create a transparent surface
        transparency = 0 #100% transparency value(0 = fully transparent, 255 = no transparency)
        transparent_surface = pygame.Surface((self.rect_width, self.rect_height), pygame.SRCALPHA)
        transparent_surface.set_alpha(transparency) #set transparency level

        #draw the rectangle on the transparent surface 
        pygame.draw.rect(transparent_surface, blue, self.ghost_rect) 

        #ghost sprite
        pygame.draw.circle(self.screen, self.color, (self.posx, self.posy), self.radius)

    def move(self):
        #movement of ghosts
        pass
             


class Pacman:
    '''Class to create pacman sprite'''

    def __init__(self, screen, posx, posy, lives = 3, radius = 25, facing=0, dotcount=0):
        self.posx = posx            # pacman's x position
        self.posy = posy            # pacman's y position
        self.facing = facing        # the direction pacman is facing[0=right, 1=left, 2 = up, 3 = down]
        self.dotcount = dotcount    # number of dots remaining to complete the level
        self.speed = 100             # speed of pacman
        self.screen = screen        # window pacman is drawn on
        self.radius = radius        # radius of pacman
        self.lives = lives          # number of lives pacman has

        self.color = PACMAN_YELLOW  #color of pacman

        self.is_dead = False        #if pacman is dead => True, otherwise => False

        #these are colors for a temporary 'death animation'
        self.death_anim_col = [red, green, blue, yellow, black, PACMAN_YELLOW]

        #get the dimensions of window
        self.width_window, self.height_window = self.screen.get_size()

        #delta_time adjusts movement speed so it isn’t affected by FPS drops
        self.delta_time = clock.tick(60) / 1000

        
        
    def draw(self):
        '''
        Draws pacman onto the window
        '''

        #find width and height of collision rect for pacman
        self.width_rect, self.height_rect = 2 * self.radius, 2 * self.radius 

        #create a collision rectangle for pacman
        self.pacman_rect = pygame.Rect(self.posx - self.radius, self.posy - self.radius, self.width_rect, self.height_rect)

        #create a transparent surface 
        transparent = 0 #100% transparency value(0 = fully transparent, 255 = opaque(0% transparent))
        transparent_surface = pygame.Surface((self.width_rect, self.height_rect), pygame.SRCALPHA)
        transparent_surface.set_alpha(transparent)
        
        #draw the rectangle on the transparent surface
        pygame.draw.rect(transparent_surface, red, self.pacman_rect)
        
        #parameters: (surface, color, center, radius)
        pygame.draw.circle(self.screen, self.color, (self.posx, self.posy), self.radius)

    def move(self):
        '''
        checks to see if the user presses any valid movement keys(arrow keys or awsd keys)
        moves pacman in the direction specified by the key pressed
        also keeps pacman from going outside of the window
        '''


        #get the key the user pressed
        key = pygame.key.get_pressed()

        #move left
        #self.posx > 0 ensures pacman does not go outside left border
        if (key[pygame.K_a] or key[pygame.K_LEFT]) and self.posx > 15:
            self.posx -= self.speed * self.delta_time

        #move right
        #can't go past right border of window
        elif (key[pygame.K_d] or key[pygame.K_RIGHT]) and self.posx < self.width_window - 15:
            self.posx += self.speed * self.delta_time
        
        #move up
        #can't go past top border of window
        elif (key[pygame.K_w] or key[pygame.K_UP]) and self.posy > 15:
            self.posy -= self.speed * self.delta_time
        
        #move down
        #can't go past bottom border of window
        elif (key[pygame.K_s] or key[pygame.K_DOWN]) and self.posy < self.height_window - 15:
            self.posy += self.speed * self.delta_time
                
    
    def check_collision_ghost(self, ghost_rect):
        ''' 
        parameters: 
        1. ghost_rect -> ghost class attribute that represents the collision rectangle that is drawn around the ghost object

        Checks if pacman collides with the ghost's collision rectangle, and if it does, it returns True. Otherwise, it returns False
        '''
        if self.pacman_rect.colliderect(ghost_rect):
            #if collision occurs, return True
            return True
        
        return False

    def dead(self):
        '''
        checks if pacman has more than 0 lives(can be revived)

        decreases the number of lives by 1

        sets boolean is_dead to True
        
        shows a color animation to inform the user that pacman has lost a life

        if pacman has 0 lives, it cannot be revived
        '''
        if self.lives > 0:
            self.lives -= 1 #lose a life
            self.is_dead = True #for debugging purposes
            print(f"Pacman has collided with a ghost. Pacman has {self.lives} hearts left")

            #death animation here:
            for col in self.death_anim_col:
                self.color = col
                self.draw() #redraw pacman with new color
                pygame.display.update() #update the screen 
                pygame.time.delay(125) #0.125 second delay between each color change

            #death sound effect here:

        else:

            print("Pacman has died!")
            #Losing page here

    def revive(self):
        '''
        revive pacman if he has at least one life left

        NOTE: the 'or True' is temporary
        '''
        if self.lives > 0 or True:
            self.is_dead = False #set to False again
            #put pacman back in default position
            self.posx, self.posy = self.width_window // 2, self.height_window // 2
            print(f"Pacman has been revived and has {self.lives} hearts")



    def check_collisions_wall(self):
        '''
        For collisions with the map, I plan on utilizing the 2d grid and using a series of if statements
        to check if pacman can turn in a certain direction and proceed with a movement without colliding with the walls 
        '''
        pass
```
